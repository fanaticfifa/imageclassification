```python
import pandas as pd
import matplotlib.pyplot as plt
from skimage.io import imread
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,ConfusionMatrixDisplay
from skimage.transform import resize
```

```python
data_array=[]
target_array=[]
categories=['Cat','Dog']
datadir='/content/drive/MyDrive/Data'
```

```python
for i in categories:
  sub_path=os.path.join(datadir,i)
  print(sub_path)
  images=os.listdir(sub_path)
  for img in images:
    img_path=os.path.join(sub_path,img)
    img_array=imread(img_path)
    img_resized=resize(img_array,(150,150,3))
    data_array.append(img_resized.flatten())
    target_array.append(categories.index(i))
  print('Loaded Category :',i)
```

```python
import numpy as np
data=np.array(data_array)
```

```python
data.shape
```

```python
target=np.array(target_array)
target.shape
```

```python
df=pd.DataFrame(data)
df
```

```python
df['target']=target
```

```python
df
```

```python
features=df.iloc[:, :-1]
target=df.iloc[:, -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=4)
```

```python
sv = SVC()
sv.fit(X_train, y_train)
y_predsvm = sv.predict(X_test)
y_predsvm
accuracy = accuracy_score(y_test, y_predsvm)
print("Accuracy:", accuracy)
```

```python
print(classification_report(y_test,y_predsvm))
print(ConfusionMatrixDisplay.from_predictions(y_test,y_predsvm))
```
